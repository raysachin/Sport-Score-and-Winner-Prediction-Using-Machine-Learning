**Now we have to extract out runs_left, balls_left, wicket_left, crr and rrr etc**

**'total_runs_y' column gives the run after each ball so we find out the score after each ball by adding previous run scored on each ball**

**for this we apply a function cumalitive sum on 'total_run_x' column**

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df['current_score']
```

**Now we add a new column to find run left after each delivery**

**Runs_left = total_runs - current_runs**

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

**Now we have to find balls_left after each delivery**

```python
delivery_df['balls_left'] = 120 - ((delivery_df['over']-1)*6 + delivery_df['ball'])
```

**Now we have to find wicket left, for this there is no direct dismissiol in int case**

**so the row NaN gives us no dismissal and row with player name gives us dismissal**

**so we replace NaN by 0 and other by 1**

**Then convert into int and then apply cumsum()**

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
```

```python
delivery_df
```

**convert into int**

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df['wicket_left'] = 10 - wickets
```

```python
delivery_df
```

**To find current run_rate**

crr = runs/over_played

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

**To find required run_rate**

rrr = runs left/ over left

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

**Results in the terms of 0 and 1**

**Logic = if winner_team == batting_team then 1 else 0**

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis = 1)
```

# Model building

**Divide into independent and dependent variable**

```python
X = final_df.iloc[:, :-1]
Y = final_df.iloc[:, -1]
```

**split into training set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 1)
```

**Here three column batting_team, bowling_team and city are in string form it will create problen so we apply one-Hot-encoding**

**One-Hot-encoding**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse = False, drop = 'first'), ['batting_team', 'bowling_team', 'city'])
], remainder = 'passthrough')
```

**Now we make a pipeline in which **

**1. Transformation**

**2. Apply Logistic regression**

**Import Logistic regression and pipeline**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

**Create a pipe object**

```python
pipe_ipl = Pipeline(steps = [
    ('step1', trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

**Feed the data**

```python
pipe_ipl.fit(X_train, Y_train)
```

**Predict the value on testing dataset**

```python
Y_pred = pipe_ipl.predict(X_test)
```

**Prediction on training dataset**

**This will give probabilty of winning and lossiing a match at the end of every over**

```python
def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['total_runs_x']))
    
```

```python
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wicket_left','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wicket_left'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
    
```

```python
temp_df,target = match_progression(delivery_df,74,pipe_ipl)
temp_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,4))
plt.plot(temp_df['end_of_over'],temp_df['wickets_in_over'],color='yellow',linewidth=3)
plt.plot(temp_df['end_of_over'],temp_df['win'],color='#00a65a',linewidth=4)
plt.plot(temp_df['end_of_over'],temp_df['lose'],color='red',linewidth=4)
plt.bar(temp_df['end_of_over'],temp_df['runs_after_over'])
plt.title('Target-' + str(target))
```
